###Project on Backing up a folder in a zip file the name of which auto increments

```python
#Generating the backup of the folder in a zip file with functionality of auto increment name

import zipfile, os
#function which can be reuse based upon the requirements
def backuptozip(folder):
  #backing up the entire content of the folder
  
  #making sure that provided folder has an absolute path
  folder = os.path.abspath(folder)

  #figuring out what fileno should be used based on the backup files already exist
  no = 1
  while True:
    zipfilename = os.path.basename(folder) +  '_' + str(no) + '.zip'
    if not os.path.exists(zipfilename):
      break
    no = no +1
  
  #now creating a zip file
  print('creating the %s' %(zipfilename))
  backupzip = zipfile.ZipFile(zipfilename, 'w')

  #no walking the entire tree in that folder and backing up every file
  for foldername, subfolders, filenames in os.walk(folder):
    print('Adding files in %s...' %(foldername))
    #adding the current folder to zipfile
    backupzip.write(foldername)
    #adding all files of current folder to zip
    for filename in filenames:
      #checking and excluding the zip files or previous archived
      newbase = os.path.basename(folder)
      if filename.startswith(newbase) and filename.endswith('.zip'):
        continue
      #backing non zip files
      backupzip.write(os.path.join(foldername,filename))
  backupzip.close()
  print('Backup Done!!')

#Calling the function for backing up sample data

backuptozip('/content/sample_data')
```

###Project on renaming the files in american date mm-dd-yyyy to europian date format dd-mm-yyyy

```python
#Code for changing dates in files name to eu format from us format

import shutil, os, re

#creating a file for testing
fileptr = open('hello_world 12-01-2019 created.txt','w')
fileptr.write('Hello World!')
fileptr.close()

#creating a regex for matching files with american date format

datepattern = re.compile(r"""
^(.*?)            #All the text before the date
((0|1)?\d)-       #one or two digits for the month
((0|1|2|3)?\d)-   #one or two digits for the day
((19|20)\d\d)     #four digits for the year
(.*?)$            #All the text after the date
""", re.VERBOSE)

#loop over the files in the working directory
for amerfilename in os.listdir('.'):
  mo = datepattern.search(amerfilename)
  #skipping a file without the date
  if mo == None:
    continue
  #get the different parts of the file name
  beforepart  = mo.group(1)
  monthpart   = mo.group(2)
  daypart     = mo.group(4)
  yearpart    = mo.group(6)
  afterpart   = mo.group(8)
  
  #From the New filename and rename files 
  eurofilename = beforepart + daypart + '-' + monthpart + '-' + yearpart + afterpart

  #get the full, absolute file paths.
  absdir = os.path.abspath('.')
  amerfilename = os.path.join(absdir, amerfilename)
  eurofilename = os.path.join(absdir, eurofilename)

  #Rename the files
  print('Reanming the "%s" to "%s" ....' %(amerfilename,eurofilename))
  shutil.move(amerfilename, eurofilename)
```

###Project: finding the all email and phone no from copied Text

```python
'''
#Project Objectives
 Get the text off the clipboard and take input
 Find all phone numbers and email addresses in the text.
 Paste them onto the clipboard.
Now you can start thinking about how this might work in code. The
code will need to do the following:
 Use the input and print module to copy and paste strings.
 Create two regexes, one for matching phone numbers and the other for
matching email addresses.
 Find all matches, not just the first match, of both regexes.
 Neatly format the matched strings into a single string to paste.
 Display some kind of message if no matches were found in the text.
'''

import re


text = input('Kindly copy and paste the text you want to extract email and phone no of')

#Creating regex for finding all phone no in the inputed text.

phonereg =  re.compile(r'''(
  (\d{3}|\(\d{3}\))?              #finding Area code
  (\s|-|\.)?                      #finding seperator
  (\d{3})                         #finding first three digit of no
  (\s|-|\.)                       #finding seperator
  (\d{4})                         #finding the last four digits of no
  (\s*(ext|x|ext.)\s*(\d{2.5}))?  #finding the extension no for the user
)''', re.I|re.VERBOSE)

#Crating regex for the finding all email address 

emailreg = re.compile(r'''(
  [a-zA-Z0-9._%+-]+               #username
  @                               # @ symbol for the email
  [a-zA-Z0-9.-]+                  #domain name
  (\.[a-zA-z]{2,4})               #dot something
)''',re.I|re.VERBOSE)

#Finding matches in the inputted text

matches = []
for groups in phonereg.findall(text):
  phoneno = '-'.join([groups[1],groups[3],groups[5]])
  if groups[8] != '':
    phoneno =+ ' x' + groups[8]
  matches.append(phoneno)
for groups in emailreg.findall(text):
  matches.append(groups[0])

#Checking if matches has been found or not

if len(matches) > 0:
  print('Matches are found and are being printed like no-newline-emailid:')
  print('\n'.join(matches))
else:
  print('No email or phone no has been found in the inserted text')

#phonereg.search('My phone no is 945 253 1884 ext 884')
#emailreg.search('My email is anubhavkumar.995@gmail.com')
#emailreg.search('My email is akumar14@ec.iitr.ac.in')

```

```python
import re
#phonenoreg = re.compile(r'\d\d\d-\d\d\d-\d\d\d\d') #r'' passes raw string
#phonenoreg = re.compile('\\d\\d\\d-\\d\\d\\d-\\d\\d\\d\\d') #other method with escape charecter
#phonenoreg = re.compile(r'\d{3}-\d{3}-\d{4}') # most shortest method
phonenoreg = re.compile(r'(\d{3})-(\d{3}-\d{4})') # most shortest method

mo = phonenoreg.search ('My Phone no is 945-253-1884 call me')
#print('Phone No found is: ' + mo.group())
print ('Found No is: ' + 'Area code: ' + mo.group(1) + ' No: ' + mo.group(2) )
# groups results in the tuple information of the all groups
areaCode, phNo = mo.groups()
print ('Area code: ' + areaCode + ' Phone No: ' + phNo )

#Searching for )( in the regex just escape it by \

reg2 = re.compile(r'(\(\d{3}\))-(\d{3}-\d{4})')
mo1 = reg2.search('My mobile no is (945)-234-1884 call me')
areaCode, phNo = mo1.groups()
#mo1.group(1)
print ('Area code: ' + areaCode + ' Phone No: ' + phNo )

#----------Using Pipe for maultiple matching or OR-ing the match---------------#
reg3 = re.compile(r'Sonu|Anubhav Kumar')
mo2 = reg3.search('My name is Anubhav Kumar. At home people call me Sonu')
mo2.group() #'Anubhav Kumar'

#------in pipe also you can group multiple words--------------#
reg4 = re.compile(r'red(wine|beer|whiskey|malt)')
mo3 = reg4.search('I love to have redwine but redwhiskey will also do')
mo3.group() #'redwine'
mo3.group(1) #'wine'

#------------Optional matching with question mark-------------#
reg4 = re.compile(r'red(-| |_)?(wine|beer|whiskey|malt)')
mo3 = reg4.search('I love to have red wine but redwhiskey will also do')
mo3.group() #'red wine'
mo3.group(2) #'wine'
mo3 = reg4.search('I love to have redwhiskey but red-wine will also do')
mo3.group() #'redwhiskey'
mo3.group(2) #'whiskey'

#-------------Matching any or no occurance of a group with *----------#
reg4 = re.compile(r'(red)*(wine|beer|whiskey|malt)')
reg5 = re.compile(r' (red)*(beer|whiskey|malt)')
mo3 = reg4.search('I love to have redredredwine but whiskey will also do')
mo3.group() #'redredredwine'
mo3.group(1) #'red'
mo3 = reg5.search('I love to have redredredwine but whiskey will also do')
mo3.group() #'whiskey'
mo3.group(1) #''

#-----------Mathing one or more occurance of a group with +---------#
reg4 = re.compile(r'(red)+(wine|beer|whiskey|malt)')
reg5 = re.compile(r' (red)+(beer|whiskey|malt)')
mo3 = reg4.search('I love to have redredredwine but whiskey will also do')
mo3.group() #'redredredwine'
mo3.group(1) #'red'
mo3 = reg5.search('I love to have redredredwine but whiskey will also do')
#mo3.group() #'None' Error
#mo3.group(1) #'None' Error
mo3 == None #'True'

#----------Using {} for no range occurance ---------------------#
#pattern will search for ha occurance in 2 or 4 time range but.
#But it is GREEDY in nature and will look for the largest match
reg = re.compile(r'(ha){2,4}')
#search1
mo = reg.search('hahaha')
mo.group() #'hahaha'
#search2
mo = reg.search('ha')
mo == None #'True'

# NON-GREEDY version for the Shortest match
reg = re.compile(r'(ha){2,4}?')
#search1
mo = reg.search('hahaha')
mo.group() #'haha'

#----------findall() function for all results/tuples for groups-----#

reg4 = re.compile(r'redwine|redwhiskey')
mo3 = reg4.findall('I love to have redwine but redwhiskey will also do')
mo3 
#o/p_noGroups: all matches: ['redwine', 'redwhiskey']

reg4 = re.compile(r'(red)(wine|beer|whiskey|malt)')
reg4.findall('I love to have redwine but redwhiskey will also do')
#o/p tuples:  [('red', 'wine'), ('red', 'whiskey')]

#-----------regex charecter classes similar to \d-------------#
#\d: 0-9 not\d: \D
#\w: [0-9,ch,_] not\w: \W
#\s: [spcae , tab\t, new-line\n] not\s: \S

reg = re.compile(r'\d+\s\w+')
reg.findall('12 drummers, 11 pipers, 10 lords, 9 ladies, 8 maids, 7 swans, 6 geese, 5 rings, 4 birds, 3 hens, 2 doves, 1 partridge')
# ['12 drummers', '11 pipers',......]

#----------Making your own charecter classes-------------------#
#Define your class in [a-zA-Z0-9] will match char(upper+lower)+numeric
reg = re.compile(r'[anubhavAnubhav0-9.@gmailcom.*?]')
reg.findall('Is My email id anubhavkumar.995@gmail.com.? *')
#o/p 1: all the charecters mentioned in the regex expression

#Defining a negative class of the above example
reg = re.compile(r'[^anubhavAnubhav0-9.@gmailcom.*?]')
reg.findall('Is My email id anubhavkumar.995@gmail.com.? *')
#o/p 2: remaining charecters of the output from 1

#--------searching the begin or end of string with ^,$ respectivly------#
#if both the wild charecters ^$(carrot cost dollars) are used then full string match is required. 
reg = re.compile(r'^\w+[a-zA-Z _\-]?\d+$')
reg.findall('My-9452531884') #match
#full string has to be matched then only search result will be there.
reg = re.compile(r'^\w+\d+$')
reg.findall('My_9452531884') #match
reg.findall('My -9452531884') #no match as \w doesn't include the space and -.
#Also keep in mind that space is not included into the \w so only word in matched
#------------------------Wild card charecter .(DOT)---------------------------#
#search multiple values of the -at word
reg = re.compile(r'.at')
reg1 = re.compile(r'\.at')
reg.findall('cat bat hat mat rat sat lal peep we .at')
#o/p: ['cat', 'bat', 'hat', 'mat', 'rat', 'sat']
reg1.findall('cat bat hat mat rat sat lal peep we .at')
#o/p: ['.at']

#with force first and last charecter limit
reg = re.compile(r'^\w.+\d+$')
reg.findall('My-9452531884') # match
reg.findall('- My-9452531884') # no match as - is not in \w match
reg.findall('My-9452531884 anubhav') # no match should end with number

##Matching anything except \n(newline) within a pattern using (.*)

#Greedy method of matching by (.*)
reg = re.compile(r'<Name: (.*)>')
#Non-Greedy method of matching by (.*?)
reg1 = re.compile(r'<Name: (.*?)>')
reg.search('<Name: Anubhav> Kumar>')# Greedy longest match <Name: Anubhav> Kumar>
reg1.findall('<Name: Anubhav> Kumar>')# Non-Greddy shortest match <Name: Anubhav>

##Matching newline with the wild card charecter .
reg = re.compile(r'.*')
reg1 = re.compile(r'.*', re.DOTALL)
reg.search('Name: Anubhav\nID: 18534002')# Name: Anubhav match till newline
reg1.search('Name: Anubhav\nID: 18534002').group()#  matches all and group gives only relevent information
reg.findall('Name: Anubhav\nID: 18534002')# findall matchs all and result in tuple

#----------Case in-sensitive regex searching with extra argument re.I---------#
reg = re.compile(r'anubhav')
reg1 = re.compile(r'anubhav',re.I)

reg.findall('I am Anubhav')# no match as match is case sensitive
reg1.findall('I am Anubhav')# "Anubhav" match as match is case in-sensitive

#-----------find and Substitue a string in place of other string---------------#
reg = re.compile(r'anubhav\s\w+',re.I)#A (.*) will include everything all words
reg.sub('Chu', 'My name is Anubhav Kumar')# 'My name is chu'

#replacing/substituting only some part of a searched string
reg = re.compile(r'(a)nubhav\s(\w)\w+',re.I)#A (.*) will include everything all words
reg.sub(r'\1** \2**', 'My name is Anubhav Kumar')# 'My name is A** K**'
reg = re.compile(r'Mr. (\w)\w+')
reg.sub(r'\1**','I love to meet Mr. Dinesh, Mr. Lomash and Mr. Mahipal')
#o/p: 'I love to meet D**, L** and M**'


#------------Organize a long regex expression using re.VERBOSE and '''--------#
#better visualization
reg = re.compile(r'''(
  (\d{3}|\(\d{3}\))? #areaCode either like 123 (123) or not at all
  (\s|-|\.)? #finding space or - or dot
  \d{3} #first three digits of Phno
  (\s|-|\.)? #finding space or - or dot
  \d{4} #finding last 4 digit of phone no;
  (\s*(ext|x|ext.)\s*\d{2,5})? #finding wether an extension no id there
)''',re.VERBOSE)
reg.search('My phone no is 945 253 1884 ext 884')
#O/P:  match='945 253 1884 ext 884'

#----------Passing one or more than one parameters in re.compile function------#
reg = re.compile(r'''(
  (a)            #Matching first a/A
  nubhav\s       #mathing the remaining part of the name
  (\w)?          #Searching for K in Kumar 
  (\w*)           #Searching the remaining part of Kumar
  )''',re.I|re.DOTALL|re.VERBOSE)#case in-sensitive|search after \n|commenting inside regex)
reg.search('My name is Anubhav/n But you can just call me Anubhav Kumar')
#O/P: 'Anubhav Kumar'
```
